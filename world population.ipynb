## Top 10 contries of total population in a year 2025-2015

```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv(
    "world_population/data/API_SP.POP.TOTL_DS2_EN_csv_v2_3107.csv",
    skiprows=4
)

# Drop unnecessary columns
df = df.drop(columns=["Indicator Name", "Indicator Code"])

# Select the latest year (change to "2023" if your dataset doesn't have 2024)
year = "2024"

# Remove rows with missing values
df = df.dropna(subset=[year])

# Remove aggregate regions (keep only countries)
aggregates = [
    "World", "High income", "Low income", "Middle income",
    "European Union", "OECD members"
]

df = df[~df["Country Name"].isin(aggregates)]

# Get the top 10 countries by population
top10 = df.sort_values(by=year, ascending=False).head(10)

# Create the bar plot
plt.figure(figsize=(12, 6))
plt.bar(top10["Country Name"], top10[year])

# Add title and axis labels
plt.title(f"Top 10 Countries by Population ({year})")
plt.xlabel("Countries")
plt.ylabel("Total Population")

# Rotate country names for better visibility
plt.xticks(rotation=45)

# Show the plot
plt.tight_layout()
plt.show()
```

```python


import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv(
    "world_population/data/API_SP.POP.TOTL_DS2_EN_csv_v2_3107.csv",
    skiprows=4
)

# Remove unnecessary columns
df = df.drop(columns=["Indicator Name", "Indicator Code"])

# Remove missing values
df = df.dropna(subset=["2015", "2024"])

# Remove aggregate regions
aggregates = [
    "World",
    "High income",
    "Low income",
    "Middle income",
    "European Union",
    "OECD members"
]

df = df[~df["Country Name"].isin(aggregates)]

# Select Top 10 countries based on 2024 population
top10 = df.sort_values(by="2024", ascending=False).head(10)

# Plot stacked bar chart
plt.figure(figsize=(14,7))

plt.bar(
    top10["Country Name"],
    top10["2015"],
    label="2015"
)

plt.bar(
    top10["Country Name"],
    top10["2024"],
    bottom=top10["2015"],
    label="2024"
)

plt.title("Top 10 Countries by Population (2015 and 2024)")
plt.xlabel("Country Name")
plt.ylabel("Population")
plt.xticks(rotation=45)

plt.legend()

plt.tight_layout()
plt.show()
```

```python
# ============================================
# Bar Chart for Mean, Median and Standard Deviation
# ============================================

import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv(
    "world_population/data/API_SP.POP.TOTL_DS2_EN_csv_v2_3107.csv",
    skiprows=4
)

# Drop unnecessary columns
df = df.drop(columns=["Indicator Name", "Indicator Code"])

# Select year
year = "2024"

# Remove missing values
df = df.dropna(subset=[year])

# Calculate statistics
mean_population = df[year].mean()
median_population = df[year].median()
std_population = df[year].std()

# Create labels and values
statistics = ["Mean", "Median", "Standard Deviation"]
values = [mean_population, median_population, std_population]

# Plot bar chart
plt.figure(figsize=(8,6))

plt.bar(statistics, values)

plt.title("Population Statistics (2024)")
plt.xlabel("Statistics")
plt.ylabel("Population")

# Display values on top of bars
for i, value in enumerate(values):
    plt.text(i, value, f"{value:,.0f}", ha="center", va="bottom")

plt.tight_layout()
plt.show()
```

## Top 10 Countries with Lowest Population Growth

```python
# ==========================================
# Top 10 Countries with Lowest Population Growth
# (2015 - 2025)
# ==========================================

import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("world population.csv")

# Calculate population growth
df["Population Growth"] = df["2025 Population"] - df["2015 Population"]

# Select the 10 countries with the lowest population growth
lowest_growth = df.nsmallest(10, "Population Growth")

# Display the results
print("Top 10 Countries with Lowest Population Growth (2015–2025)")
print(lowest_growth[["Country", "2015 Population", "2025 Population", "Population Growth"]])

# Plot the graph
plt.figure(figsize=(12,6))
plt.bar(lowest_growth["Country"], lowest_growth["Population Growth"])

plt.title("Top 10 Countries with Lowest Population Growth (2015–2025)")
plt.xlabel("Country")
plt.ylabel("Population Growth")

plt.xticks(rotation=45)
plt.grid(axis="y")
plt.tight_layout()

plt.show()
```
